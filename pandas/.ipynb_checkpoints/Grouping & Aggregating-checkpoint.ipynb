# pandas.DataFrame.groupby
Group DataFrame using a mapper or by a Series of columns.
A groupby operation involves some combination of  splitting the object , applying a function,
and combining the results. This can be used to group large amounts of data and compute operations on these groups.

# Grouping 

```python
country_grp = df.groupby(['Country'])
```

```python
uu = country_grp.get_group('United States of America')
```

```python
filt = df['Country'] == 'United States of America'
```

```python
df.loc[filt]['LanguageWantToWorkWith'].value_counts()
```

```python
df['LanguageWantToWorkWith'].value_counts()
```

```python
country_grp['LanguageWantToWorkWith'].value_counts(normalize=True).loc['United States of America']
```

```python
country_grp['LanguageWantToWorkWith'].value_counts().loc['United States of America']
```

```python
country_grp['LanguageWantToWorkWith'].value_counts().loc['China']
```

```python
country_grp['CompTotal'].median().loc['China']
```

```python
country_grp['CompTotal'].agg(['median','mean']).tail(50)
```

```python
country_grp['CompTotal'].agg(['median','mean']).loc['United States of America']
```

```python
country_grp['LanguageWantToWorkWith'].apply(lambda x : x.str.contains('Python').sum())
```

```python
country_respondents = df['Country'].value_counts()
```

```python
country_respondents
```

```python
country_user_python = country_grp['LanguageWantToWorkWith'].apply(lambda x : x.str.contains('Python').sum())
country_user_python
```

# Combining

```python
python_df = pd.concat([country_user_python,country_respondents],axis = 'columns',sort=False)
```
