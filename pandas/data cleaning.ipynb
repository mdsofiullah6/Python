# Data Cleaning 
Data cleaning means fixing bad data in your data set.

# Bad data could be:
1: Empty cells

2: Data in wrong format

3: Wrong Data

4: Duplicates


1; Empty cells

Empty cells can potentially give you a wrong result when you analyze data.

# Remove Empty Rows

There is only way to deal with empty cells to remove rows that contains empty cell.
This is usuallyl ok . since data sets can be very big , and removing a few rows will not have big impect on the result.

```python
import pandas as pd 
import numpy as np

sf = pd.read_csv('data/survey_results_public.csv')

```

```python
people = {
    'name' : ['richard', 'dahyun' , 'nayeon' , 'sana',''],
    'title' : ['singularity','kim', np.nan , None,''],
    'country' : ['united state' , 'south korea' , 'south korea', 'japan' ,'Mising'],
    'food' : ['meat' , 'milk,chocklet milk' , 'salmon sushi',np.nan,None],
    'email' : ['richard@email.com' , 'dahyun@email.com' , 'nayeon@email.com', 'sana@email.com',''] ,
    'height_in_feet' : [np.nan , '5.3' , '5.4', '5.4' ,''],
    'age' : ['20' , '24' , '27' ,'24','']
}
df = pd.DataFrame(people)
```
