```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("fiverr_victorioustudio.csv")

# Apply the value_counts method to the username column
counts = df['username'].value_counts()

# Convert the result of value_counts to a DataFrame
counts = counts.to_frame().reset_index()
```

```python
counts
```

```python
counts.to_csv('rank.csv')
```

```python
# Rename the columns for clarity
counts.columns = ['username', 'count']

# Merge the counts DataFrame with the original DataFrame on the username column
result = counts.merge(df, on='username', how='left').drop_duplicates(subset='username')

# Rearrange the columns as desired
result = result[['username', 'country-name', 'count']]
```

```python
result = result.reset_index(drop=True)
```
