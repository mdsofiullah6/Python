# Filtering Data

```python
import pandas as pd 
import numpy as np
```

```python
ff = {
    'name' : ['richard', 'dahyun' , 'nayeon' , 'sana'],
    'country' : ['united state' , 'south korea' , 'south korea', 'japan' ],
    'food' : ['meat' , 'chocklet milk' , 'salmon sushi','spicy food'],
    'email' : ['richard@email.com' , 'dahyun@email.com' , 'nayeon@email.com', 'sana@email.com'] ,
    'height_in_feet' : ['5.7' , '5.3' , '5.4', '5.4' ],
    'age' : ['18' , '24' , '27' ,'24']
}
df = pd.DataFrame(ff)
```

```python
fil = (df['age'] == '24') & ( df['name'] == 'dahyun' )
```

```python
df.loc[fil, 'email']
```

# | 

we use bar [| ] for multiple filtering .

```python
fil1 = (df['age'] == '24') | ( df['country'] == 'united state' )

df.loc[fil1,'email']
```
